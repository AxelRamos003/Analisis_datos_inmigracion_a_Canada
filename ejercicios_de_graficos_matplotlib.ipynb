```python
import pandas as pd
```

```python
archivo = '/content/inmigrantes_canada.csv'
```

```python
df_datos= pd.read_csv(archivo)
df_datos.head(4)
```

 crear un gráfico de líneas comparando los números de inmigrantes de Brasil y Argentina a Canadá

```python
df_datos= df_datos.set_index('Pais')
df_datos.head(4)
```

```python
anos = list(map(str, range(1980,2014)))
```

```python
df_datos_Argentina = df_datos.loc['Argentina',anos]
df_datos_Argentina.head(4)
```

```python
dict_argentina = {'Anos': df_datos_Argentina.index.tolist(),
                  'Inmigrantes': df_datos_Argentina.values.tolist()}
dict_argentina
```

```python
Inmigrantes_Argentina= pd.DataFrame(dict_argentina)
Inmigrantes_Argentina.head(3)
```

```python
df_datos_Brasil = df_datos.loc['Brasil',anos]
df_datos_Brasil.head(4)
```

```python
dict_brasil = {'Anos': df_datos_Brasil.index.tolist(),
                  'Inmigrantes': df_datos_Brasil.values.tolist()}
```

```python
Inmigrantes_Brasil= pd.DataFrame(dict_brasil)
Inmigrantes_Brasil.head(3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Inmigrantes_Argentina['Anos'],Inmigrantes_Argentina['Inmigrantes'])
plt.plot(Inmigrantes_Brasil['Anos'],Inmigrantes_Brasil['Inmigrantes'])
plt.legend(['Argentina','Brasil'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.xlabel('Anos')
plt.ylabel('Inmigrantes')
plt.show()
```
